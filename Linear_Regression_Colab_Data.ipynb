```python
from sklearn.linear_model import LinearRegression

df_x = df.drop(labels=["median_house_value"], axis=1)
model = LinearRegression()
model.fit(df_x, df_y)

param_score_arr = []
```

```python
# df_x has full features (8)
model.score(df_x, df_y)
```

```python
df_x2 = df_x.drop(labels=["longitude"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['longitude', new_var])
```

```python
df_x2 = df_x.drop(labels=["latitude"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['latitude', new_var])
```

```python
df_x2 = df_x.drop(labels=["housing_median_age"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['housing_median_age', new_var])
```

```python
df_x2 = df_x.drop(labels=["total_rooms"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['total_rooms', new_var])
```

```python
df_x2 = df_x.drop(labels=["total_bedrooms"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['total_bedrooms', new_var])
```

```python
df_x2 = df_x.drop(labels=["population"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['population', new_var])
```

```python
df_x2 = df_x.drop(labels=["households"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['households', new_var])
```

```python
df_x2 = df_x.drop(labels=["median_income"], axis=1)
model.fit(df_x2, df_y)
new_var = model.score(df_x2, df_y)
print(new_var)
param_score_arr.append(['median_income', new_var])
```

```python

max_val = 0
max_char = ''

for item in param_score_arr:
    if item[1] > max_val:
        max_val = item[1]
        max_char = item[0]

print(f'Without this parameter: {max_char} -- Maximum accuracy: {max_val}')
```

```python
model1 = LinearRegression()

param_score_arr1 = []
input_parameters = ['longitude', 'latitude',	'housing_median_age',	'total_rooms', 'total_bedrooms',	'population',	'households',	'median_income']

for i in range(len(input_parameters)):
    for j in range(i+1, len(input_parameters)):
        df_x3 = df_x.drop(labels=[input_parameters[i], input_parameters[j]], axis=1)
        model1.fit(df_x3, df_y)
        new_var = model1.score(df_x3, df_y)
        param_score_arr1.append([input_parameters[i] + ' AND ' + input_parameters[j], new_var])

print(f'Total group count: {param_score_arr1.__len__()}')
max_val = 0
max_char = ''

for item in param_score_arr1:
    if item[1] > max_val:
        max_val = item[1]
        max_char = item[0]

print(f'Without this parameters: {max_char} -- Maximum accuracy: {max_val}')
```
